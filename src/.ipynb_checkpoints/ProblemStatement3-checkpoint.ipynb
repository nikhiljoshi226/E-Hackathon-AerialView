```python
import os
import time
import warnings
from typing import Dict, List

import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")

# ============================================================
# Problem 3 — Crypto Blind Anomaly Hunt
# Usage:
#   python problem3_updated.py
#
# Expected project layout:
#   data_dir/
#     crypto-market/
#       Binance_SYMBOL_2026_minute.csv
#     crypto-trades/
#       SYMBOL_trades.csv
#
# Output:
#   submission.csv
# ============================================================

DATA_DIR = "data_dir"
MARKET_DIR = os.path.join(DATA_DIR, "crypto-market")
TRADES_DIR = os.path.join(DATA_DIR, "crypto-trades")

PAIRS = [
    "USDCUSDT", "BATUSDT", "DOGEUSDT", "LTCUSDT",
    "XRPUSDT", "SOLUSDT", "ETHUSDT", "BTCUSDT",
]

ACCEPTED_VIOLATION_TYPES = {
    "aml_structuring",
    "coordinated_structuring",
    "threshold_testing",
    "chain_layering",
    "manager_consolidation",
    "placement_smurfing",
    "wash_trading",
    "pump_and_dump",
    "layering_echo",
    "spoofing",
    "ramping",
    "coordinated_pump",
    "round_trip_wash",
    "cross_pair_divergence",
    "peg_break",
    "wash_volume_at_peg",
}

THRESHOLDS = [1, 10, 50, 100, 200, 500, 1000, 5000]


def load_pair(symbol: str) -> tuple[pd.DataFrame, pd.DataFrame]:
    market_path = os.path.join(MARKET_DIR, f"Binance_{symbol}_2026_minute.csv")
    trades_path = os.path.join(TRADES_DIR, f"{symbol}_trades.csv")

    mkt = pd.read_csv(market_path)
    mkt.columns = [c.strip() for c in mkt.columns]
    mkt = mkt.rename(
        columns={
            "Date": "timestamp",
            "Open": "open",
            "High": "high",
            "Low": "low",
            "Close": "close",
        }
    )
    mkt["timestamp"] = pd.to_datetime(mkt["timestamp"])

    vol_col = [
        c for c in mkt.columns
        if c.lower().startswith("volume") and "usdt" not in c.lower()
    ]
    if vol_col:
        mkt = mkt.rename(columns={vol_col[0]: "volume"})
    elif "Volume" in mkt.columns:
        mkt = mkt.rename(columns={"Volume": "volume"})

    mkt = mkt.sort_values("timestamp").reset_index(drop=True)

    tr = pd.read_csv(trades_path)
    tr["timestamp"] = pd.to_datetime(tr["timestamp"])
    tr["date"] = tr["timestamp"].dt.strftime("%Y-%m-%d")
    tr["notional"] = tr["price"] * tr["quantity"]
    tr["symbol"] = symbol

    return mkt, tr


def safe_cv(series: pd.Series) -> float:
    mean_val = series.mean()
    if pd.isna(mean_val) or mean_val == 0:
        return 999.0
    std_val = series.std(ddof=1)
    if pd.isna(std_val):
        return 0.0
    return float(std_val / mean_val)


def monotonic_up_ratio(prices: pd.Series) -> float:
    diffs = prices.diff().dropna()
    if len(diffs) == 0:
        return 0.0
    return float((diffs > 0).mean())


def monotonic_down_ratio(prices: pd.Series) -> float:
    diffs = prices.diff().dropna()
    if len(diffs) == 0:
        return 0.0
    return float((diffs < 0).mean())


def classify_subgroup(sub: pd.DataFrame, symbol: str) -> str:
    """
    Classify one date-level trade phase within a linked campaign group.
    """
    sub = sub.sort_values("timestamp").copy()
    wallets = sub["trader_id"].dropna().unique()
    sides = sub["side"].tolist()
    n = len(sub)

    buy_q = sub.loc[sub["side"] == "BUY", "quantity"].sum()
    sell_q = sub.loc[sub["side"] == "SELL", "quantity"].sum()
    total_q = buy_q + sell_q
    net_ratio = abs(buy_q - sell_q) / total_q if total_q > 0 else 0.0
    qty_cv = safe_cv(sub["quantity"])

    if symbol == "USDCUSDT":
        if (sub["price"].sub(1.0).abs() > 0.005).any():
            return "peg_break"
        return "wash_volume_at_peg"

    # Single-wallet patterns
    if len(wallets) == 1:
        wallet_side_buy_ratio = float((sub["side"] == "BUY").mean())
        prices = sub["price"]
        up_ratio = monotonic_up_ratio(prices)

        if n >= 4 and wallet_side_buy_ratio >= 0.80 and up_ratio >= 0.75 and prices.iloc[-1] > prices.iloc[0]:
            return "ramping"

        if n >= 4:
            buy_count = sides.count("BUY")
            sell_count = sides.count("SELL")
            if (buy_count >= 3 and sell_count >= 1 and sides[-1] == "SELL") or (
                sell_count >= 3 and buy_count >= 1 and sides[-1] == "BUY"
            ):
                return "layering_echo"

        if n >= 3:
            first_notional = float(sub["notional"].iloc[0])
            rest_notionals = sub["notional"].iloc[1:]
            for thresh in THRESHOLDS:
                if thresh <= 0:
                    continue
                if abs(first_notional - thresh) / thresh < 0.005 and (rest_notionals < thresh * 0.99).all():
                    return "threshold_testing"

        if n >= 3 and wallet_side_buy_ratio >= 0.80 and qty_cv < 0.40 and sub["notional"].mean() < 30:
            return "placement_smurfing"

        if n >= 3 and wallet_side_buy_ratio >= 0.70 and qty_cv < 0.15:
            return "aml_structuring"

    # Two-wallet / multi-wallet patterns
    if len(wallets) == 2:
        prices = sub["price"].tolist()
        alternating_sides = all(sides[i] != sides[i + 1] for i in range(len(sides) - 1)) if len(sides) >= 2 else False
        non_decreasing_steps = sum(prices[i] <= prices[i + 1] for i in range(len(prices) - 1))

        if n >= 3 and alternating_sides and non_decreasing_steps >= max(1, int(0.60 * max(1, len(prices) - 1))):
            return "chain_layering"

        if net_ratio < 0.15:
            w1, w2 = wallets[0], wallets[1]
            s1 = sub.loc[sub["trader_id"] == w1, "side"].mode()
            s2 = sub.loc[sub["trader_id"] == w2, "side"].mode()
            if not s1.empty and not s2.empty and s1.iloc[0] != s2.iloc[0]:
                return "round_trip_wash"

    wallet_buy_ratio = float((sub["side"] == "BUY").mean())
    if len(wallets) >= 2 and wallet_buy_ratio >= 0.80:
        if qty_cv < 0.25:
            return "coordinated_structuring"
        return "coordinated_pump"

    if net_ratio < 0.20:
        return "wash_trading"

    if len(wallets) >= 2 and wallet_buy_ratio >= 0.70:
        return "coordinated_structuring"

    return "aml_structuring"


def classify_group(group_df: pd.DataFrame, symbol: str) -> Dict[int, str]:
    """
    Classify all trades in one linked campaign group.
    """
    result: Dict[int, str] = {}
    group_df = group_df.sort_values("timestamp").copy()
    group_tag = group_df["manager_id"].iloc[0]

    if symbol == "USDCUSDT":
        for idx, row in group_df.iterrows():
            result[idx] = "peg_break" if abs(row["price"] - 1.0) > 0.005 else "wash_volume_at_peg"
        return result

    # Consolidation-like receipts or unusually large transfers inside the linked group.
    for idx, row in group_df.iterrows():
        if row["trader_id"] == group_tag:
            result[idx] = "manager_consolidation"

    for wallet, wallet_df in group_df.groupby("trader_id"):
        if wallet == group_tag or len(wallet_df) < 2:
            continue
        median_qty = wallet_df["quantity"].median()
        if pd.isna(median_qty) or median_qty <= 0:
            continue
        for idx, row in wallet_df.iterrows():
            if idx not in result and row["quantity"] >= median_qty * 4.0:
                result[idx] = "manager_consolidation"

    remaining = group_df.loc[~group_df.index.isin(result)].copy()
    if remaining.empty:
        return result

    for _, date_df in remaining.groupby("date"):
        vtype = classify_subgroup(date_df, symbol)
        for idx in date_df.index:
            result[idx] = vtype

    return result


def make_remark(row: pd.Series, group_df: pd.DataFrame, vtype: str, symbol: str) -> str:
    wallet = row["trader_id"]
    group_tag = row["manager_id"]
    date = row["date"]
    day_df = group_df[group_df["date"] == date].sort_values("timestamp")
    wallet_day_df = day_df[day_df["trader_id"] == wallet].sort_values("timestamp")

    if vtype == "peg_break":
        deviation_pct = abs(row["price"] - 1.0) * 100
        return (
            f"USDC trade price {row['price']:.5f} deviates {deviation_pct:.3f}% from the $1.00 peg "
            f"with qty={row['quantity']:.4f}. Linked group={group_tag}."
        )

    if vtype == "wash_volume_at_peg":
        day_buy = day_df.loc[day_df["side"] == "BUY", "quantity"].sum()
        day_sell = day_df.loc[day_df["side"] == "SELL", "quantity"].sum()
        return (
            f"Repeated linked activity at the peg on {date}; buys={day_buy:.4f}, sells={day_sell:.4f}, "
            f"wallet={wallet}, group={group_tag}."
        )

    if vtype == "manager_consolidation":
        return (
            f"Large linked receipt/transfer for wallet={wallet}; qty={row['quantity']:.4f}, "
            f"notional={row['notional']:.2f}, group={group_tag}."
        )

    if vtype == "ramping":
        p0 = wallet_day_df["price"].iloc[0] if len(wallet_day_df) else row["price"]
        p1 = wallet_day_df["price"].iloc[-1] if len(wallet_day_df) else row["price"]
        move_pct = ((p1 / p0) - 1) * 100 if p0 else 0.0
        return (
            f"Wallet={wallet} shows sequential buy pressure in {symbol} on {date}; "
            f"price moved {p0:.6f}->{p1:.6f} ({move_pct:.2f}%), group={group_tag}."
        )

    if vtype == "chain_layering":
        wallets = list(day_df["trader_id"].unique())
        return (
            f"Linked wallets {wallets} alternate sides in a stepped sequence on {date}; "
            f"wallet={wallet}, side={row['side']}, group={group_tag}."
        )

    if vtype == "coordinated_pump":
        wallets = list(day_df["trader_id"].unique())
        return (
            f"Multi-wallet buy concentration in {symbol} on {date}; wallets={wallets}, "
            f"wallet={wallet}, qty={row['quantity']:.4f}, group={group_tag}."
        )

    if vtype == "round_trip_wash":
        day_buy = day_df.loc[day_df["side"] == "BUY", "quantity"].sum()
        day_sell = day_df.loc[day_df["side"] == "SELL", "quantity"].sum()
        total_q = day_buy + day_sell
        net_ratio = abs(day_buy - day_sell) / total_q if total_q > 0 else 0.0
        return (
            f"Two-wallet mirrored activity on {date}; buys={day_buy:.4f}, sells={day_sell:.4f}, "
            f"net_ratio={net_ratio:.3f}, wallet={wallet}, group={group_tag}."
        )

    if vtype == "layering_echo":
        side_counts = wallet_day_df["side"].value_counts().to_dict()
        return (
            f"Wallet={wallet} trades mostly one direction and then reverses on {date}; "
            f"side_counts={side_counts}, group={group_tag}."
        )

    if vtype == "threshold_testing":
        first_notional = float(wallet_day_df["notional"].iloc[0]) if len(wallet_day_df) else float(row["notional"])
        return (
            f"Wallet={wallet} starts near a round-number notional ({first_notional:.2f}) and then stays below it; "
            f"date={date}, group={group_tag}."
        )

    if vtype == "placement_smurfing":
        return (
            f"Wallet={wallet} places many small linked buys with low notional dispersion on {date}; "
            f"avg_notional={wallet_day_df['notional'].mean():.2f}, group={group_tag}."
        )

    if vtype == "coordinated_structuring":
        wallets = list(day_df["trader_id"].unique())
        qty_cv = safe_cv(day_df["quantity"])
        return (
            f"Linked wallets show uniform buy sizing on {date}; wallets={wallets}, qty_cv={qty_cv:.3f}, "
            f"wallet={wallet}, group={group_tag}."
        )

    if vtype == "aml_structuring":
        qty_cv = safe_cv(wallet_day_df["quantity"])
        return (
            f"Wallet={wallet} uses repeated similar-size buys on {date}; n={len(wallet_day_df)}, "
            f"qty_cv={qty_cv:.3f}, group={group_tag}."
        )

    buy_q = wallet_day_df.loc[wallet_day_df["side"] == "BUY", "quantity"].sum()
    sell_q = wallet_day_df.loc[wallet_day_df["side"] == "SELL", "quantity"].sum()
    total_q = buy_q + sell_q
    net_ratio = abs(buy_q - sell_q) / total_q if total_q > 0 else 0.0
    return (
        f"Wallet={wallet} shows low net directional change on {date}; buys={buy_q:.4f}, sells={sell_q:.4f}, "
        f"net_ratio={net_ratio:.3f}, group={group_tag}."
    )


def build_submission() -> pd.DataFrame:
    start_time = time.time()
    all_results: List[dict] = []

    print("=" * 68)
    print("Problem 3 — Submission Builder")
    print("=" * 68)

    for symbol in PAIRS:
        print(f"\n[{symbol}]", end=" ")
        try:
            _, trades = load_pair(symbol)
        except FileNotFoundError as exc:
            print(f"missing file: {exc}")
            continue

        linked_trades = trades[trades["manager_id"].notna()].copy()
        if linked_trades.empty:
            print("0 linked campaign trades")
            continue

        print(f"{len(trades):,} total | {len(linked_trades)} linked trades | {linked_trades['manager_id'].nunique()} groups")

        symbol_rows: List[dict] = []
        for _, group_df in linked_trades.groupby("manager_id"):
            group_df = group_df.sort_values("timestamp")
            vtypes = classify_group(group_df, symbol)

            for idx, row in group_df.iterrows():
                vtype = vtypes.get(idx, "wash_trading")
                if vtype not in ACCEPTED_VIOLATION_TYPES:
                    vtype = "wash_trading"

                symbol_rows.append(
                    {
                        "symbol": symbol,
                        "date": row["date"],
                        "trade_id": row["trade_id"],
                        "violation_type": vtype,
                        "remarks": make_remark(row, group_df, vtype, symbol),
                    }
                )

        type_counts = pd.Series([r["violation_type"] for r in symbol_rows]).value_counts()
        for vtype, count in type_counts.items():
            print(f"    {vtype}: {count}")

        all_results.extend(symbol_rows)

    if not all_results:
        raise RuntimeError("No submission rows were generated. Check your DATA_DIR path.")

    sub = pd.DataFrame(all_results)
    sub = sub.drop_duplicates(subset=["trade_id"], keep="first")
    sub = sub[["symbol", "date", "trade_id", "violation_type", "remarks"]]
    sub = sub.sort_values(["symbol", "date", "trade_id"]).reset_index(drop=True)

    elapsed = time.time() - start_time
    sub.to_csv("submission.csv", index=False)

    print("\n" + "=" * 68)
    print(f"Completed in {elapsed:.2f}s | {len(sub)} flagged trades | wrote submission.csv")
    print("=" * 68)
    print("\nBy violation_type:")
    print(sub["violation_type"].value_counts().to_string())
    print("\nBy symbol:")
    print(sub["symbol"].value_counts().to_string())

    return sub


if __name__ == "__main__":
    build_submission()
```
