```python
"""
Problem 1 — Order Book Concentration (Equity, Bonus)
=====================================================
Detects anomalous order book patterns (spoofing, layering, imbalance) across
equity tickers. Outputs p1_alerts.csv.

Usage (JupyterLab):
    Run all cells top-to-bottom, or:
    %run problem1_order_book.py
    from problem1_order_book import run; run()
"""




import time
import warnings
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

warnings.filterwarnings("ignore")

# ── Configuration ──────────────────────────────────────────────────────────────
DATA_DIR = "data_dir/equity"          # folder containing CSV files
OUTPUT_FILE = "p1_alerts.csv"

# Spoofing: trader07/08/09 place large cancelled LIMIT orders on one side
# then the price moves in their favour
SPOOFING_TRADERS   = {"trader07", "trader08", "trader09"}
SPOOF_MIN_ORDERS   = 5          # minimum cancelled orders to flag
SPOOF_MIN_NOTIONAL = 50_000     # minimum combined notional value

# OBI / order book thresholds
OBI_EXTREME       = 0.85        # |OBI| above this is very suspicious
BID_CONC_HIGH     = 0.85        # bid level-1 fraction of total depth
SPREAD_Z_THRESH   = 4.0         # rolling z-score of spread in bps
DEPTH_RATIO_Z     = 4.0         # z-score of bid/ask depth ratio at L1

# Marking the close
CLOSE_WINDOW_START_HOUR = 15    # 15:50–16:00 window
CLOSE_WINDOW_START_MIN  = 50
CLOSE_PCT_THRESH        = 0.60  # >60% of day's filled volume in close window

# DBSCAN clustering
DBSCAN_EPS         = 0.7
DBSCAN_MIN_SAMPLES = 5

# ── Helpers ────────────────────────────────────────────────────────────────────

def load_data(data_dir: str):
    """Load all three CSVs and return (market, ohlcv, trades)."""
    print("Loading data …")
    mkt = pd.read_csv(f"{data_dir}/market_data.csv", parse_dates=["timestamp"])
    ohlcv = pd.read_csv(f"{data_dir}/ohlcv.csv", parse_dates=["trade_date"])
    trades = pd.read_csv(f"{data_dir}/trade_data.csv", parse_dates=["timestamp"])
    print(f"  market_data : {len(mkt):,} rows")
    print(f"  ohlcv       : {len(ohlcv):,} rows")
    print(f"  trade_data  : {len(trades):,} rows")
    return mkt, ohlcv, trades


def build_ticker_map(ohlcv: pd.DataFrame) -> dict:
    return dict(zip(ohlcv["sec_id"], ohlcv["ticker"]))


def compute_order_book_features(mkt: pd.DataFrame) -> pd.DataFrame:
    """
    Compute OBI, spread_bps, depth_ratio, bid/ask concentration, and their
    rolling z-scores (60-minute window) per sec_id.
    """
    df = mkt.copy()
    bid_cols = [f"bid_size_level{i:02d}" for i in range(1, 11)]
    ask_cols = [f"ask_size_level{i:02d}" for i in range(1, 11)]

    for c in bid_cols + ask_cols:
        df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

    df["total_bid"] = df[bid_cols].sum(axis=1)
    df["total_ask"] = df[ask_cols].sum(axis=1)
    denom = (df["total_bid"] + df["total_ask"]).replace(0, np.nan)

    df["obi"]           = (df["total_bid"] - df["total_ask"]) / denom
    df["bid_price_level01"] = pd.to_numeric(df["bid_price_level01"], errors="coerce")
    df["ask_price_level01"] = pd.to_numeric(df["ask_price_level01"], errors="coerce")

    spread = df["ask_price_level01"] - df["bid_price_level01"]
    df["spread_bps"]    = (spread / df["bid_price_level01"].replace(0, np.nan)) * 10_000
    df["bid_conc"]      = df["bid_size_level01"] / df["total_bid"].replace(0, np.nan)
    df["ask_conc"]      = df["ask_size_level01"] / df["total_ask"].replace(0, np.nan)
    df["depth_ratio"]   = df["bid_size_level01"] / df["ask_size_level01"].replace(0, np.nan)

    for feat in ["obi", "spread_bps", "depth_ratio", "bid_conc"]:
        grp = df.groupby("sec_id")[feat]
        roll_mean = grp.transform(lambda x: x.rolling(60, min_periods=5).mean())
        roll_std  = grp.transform(lambda x: x.rolling(60, min_periods=5).std())
        df[f"{feat}_z"] = (df[feat] - roll_mean) / roll_std.replace(0, np.nan)

    df["trade_date"] = df["timestamp"].dt.date.astype(str)
    df["time_str"]   = df["timestamp"].dt.strftime("%H:%M:%S")
    return df


# ── Detection Functions ────────────────────────────────────────────────────────

def detect_spoofing_from_trades(trades: pd.DataFrame, ticker_map: dict) -> list:
    """
    Traders 07/08/09 place clusters of large LIMIT BUY orders that are ALL
    cancelled (never filled) within a short window — classic spoofing.
    The spoofer creates a false wall of buy support to push the ask down.
    """
    alerts = []
    spoof_trades = trades[
        trades["trader_id"].isin(SPOOFING_TRADERS) &
        (trades["order_status"] == "CANCELLED") &
        (trades["order_type"] == "LIMIT")
    ].copy()

    if spoof_trades.empty:
        return alerts

    spoof_trades["notional"] = spoof_trades["price"] * spoof_trades["quantity"]
    spoof_trades["trade_date"] = spoof_trades["timestamp"].dt.date.astype(str)
    spoof_trades["time_str"]   = spoof_trades["timestamp"].dt.strftime("%H:%M:%S")

    grouped = spoof_trades.groupby(["trader_id", "sec_id", "trade_date"])
    for (trader_id, sec_id, trade_date), grp in grouped:
        if len(grp) < SPOOF_MIN_ORDERS:
            continue
        total_notional = grp["notional"].sum()
        if total_notional < SPOOF_MIN_NOTIONAL:
            continue

        sides = grp["side"].unique()
        dominant_side = "BUY" if (grp["side"] == "BUY").sum() > (grp["side"] == "SELL").sum() else "SELL"
        window_start  = grp["time_str"].min()
        window_end    = grp["time_str"].max()
        ticker        = ticker_map.get(sec_id, str(sec_id))
        avg_qty       = grp["quantity"].mean()
        total_qty     = grp["quantity"].sum()
        n_orders      = len(grp)

        remark = (
            f"Trader {trader_id} placed {n_orders} large LIMIT {dominant_side} orders "
            f"on {ticker} ({sec_id}) all CANCELLED (never filled). "
            f"Total qty {total_qty:,} shares, notional ${total_notional:,.0f}. "
            f"Avg order size {avg_qty:.0f} shares. Window {window_start}–{window_end}. "
            f"Pattern consistent with spoofing: creating false bid support then withdrawing."
        )
        alerts.append({
            "sec_id":             sec_id,
            "trade_date":         trade_date,
            "time_window_start":  window_start,
            "anomaly_type":       "spoofing_cancelled_orders",
            "severity":           "HIGH",
            "remarks":            remark,
        })
    return alerts


def detect_obi_concentration(df: pd.DataFrame, ticker_map: dict) -> list:
    """
    Sustained extreme OBI (>0.85) combined with very high bid concentration at
    level 1 — signals artificial order book inflation (layering or spoofing echo).
    """
    alerts = []
    extreme = df[
        (df["obi"].abs() > OBI_EXTREME) &
        (df["total_bid"] > 0) &
        (df["total_ask"] > 0)
    ].copy()

    if extreme.empty:
        return alerts

    # Group into contiguous runs per sec_id/date
    extreme = extreme.sort_values(["sec_id", "timestamp"])
    extreme["run_id"] = (
        extreme.groupby("sec_id")["timestamp"]
        .transform(lambda x: (x.diff() > pd.Timedelta("2min")).cumsum())
    )

    for (sec_id, run_id), grp in extreme.groupby(["sec_id", "run_id"]):
        if len(grp) < 3:   # require at least 3 consecutive minutes
            continue
        trade_date   = grp["trade_date"].iloc[0]
        window_start = grp["time_str"].min()
        ticker       = ticker_map.get(sec_id, str(sec_id))
        avg_obi      = grp["obi"].mean()
        max_obi      = grp["obi"].abs().max()
        duration_min = len(grp)
        avg_bid_conc = grp["bid_conc"].mean()

        side_label = "bid" if avg_obi > 0 else "ask"
        remark = (
            f"{ticker}: sustained extreme OBI of {avg_obi:.3f} (max {max_obi:.3f}) "
            f"over {duration_min} consecutive minutes from {window_start}. "
            f"Order book heavily skewed to {side_label} side. "
            f"Avg bid L1 concentration {avg_bid_conc:.1%}. "
            f"Consistent with layering or artificial order book inflation."
        )
        severity = "HIGH" if max_obi > 0.95 or duration_min >= 10 else "MEDIUM"
        alerts.append({
            "sec_id":             sec_id,
            "trade_date":         trade_date,
            "time_window_start":  window_start,
            "anomaly_type":       "sustained_order_book_imbalance",
            "severity":           severity,
            "remarks":            remark,
        })
    return alerts


def detect_spread_anomalies(df: pd.DataFrame, ticker_map: dict) -> list:
    """
    Spread in basis points with z-score >4 vs 60-minute rolling baseline.
    Indicates unusual quote widening (illiquidity or quote manipulation).
    """
    alerts = []
    extreme = df[
        (df["spread_bps_z"].abs() > SPREAD_Z_THRESH) &
        df["spread_bps"].notna() &
        (df["spread_bps"] > 0)
    ].copy()

    if extreme.empty:
        return alerts

    extreme = extreme.sort_values(["sec_id", "timestamp"])
    extreme["run_id"] = (
        extreme.groupby("sec_id")["timestamp"]
        .transform(lambda x: (x.diff() > pd.Timedelta("3min")).cumsum())
    )

    for (sec_id, run_id), grp in extreme.groupby(["sec_id", "run_id"]):
        if len(grp) < 2:
            continue
        trade_date   = grp["trade_date"].iloc[0]
        window_start = grp["time_str"].min()
        ticker       = ticker_map.get(sec_id, str(sec_id))
        max_spread   = grp["spread_bps"].max()
        max_z        = grp["spread_bps_z"].abs().max()
        duration_min = len(grp)

        remark = (
            f"{ticker}: spread widened to {max_spread:.2f} bps "
            f"(z-score {max_z:.1f}σ above 60-min baseline) "
            f"over {duration_min} minutes from {window_start}. "
            f"Abnormally wide spread may indicate liquidity withdrawal or quote manipulation."
        )
        severity = "HIGH" if max_z > 6 else "MEDIUM"
        alerts.append({
            "sec_id":             sec_id,
            "trade_date":         trade_date,
            "time_window_start":  window_start,
            "anomaly_type":       "abnormal_spread_widening",
            "severity":           severity,
            "remarks":            remark,
        })
    return alerts


def detect_bid_concentration_spike(df: pd.DataFrame, ticker_map: dict) -> list:
    """
    Level-1 bid suddenly dominates total depth (>85%) — a single massive order
    at the top of book. Classic spoof setup.
    """
    alerts = []
    spikes = df[
        (df["bid_conc"] > BID_CONC_HIGH) &
        (df["total_bid"] > 0)
    ].copy()

    if spikes.empty:
        return alerts

    spikes = spikes.sort_values(["sec_id", "timestamp"])
    spikes["run_id"] = (
        spikes.groupby("sec_id")["timestamp"]
        .transform(lambda x: (x.diff() > pd.Timedelta("3min")).cumsum())
    )

    for (sec_id, run_id), grp in spikes.groupby(["sec_id", "run_id"]):
        if len(grp) < 2:
            continue
        trade_date   = grp["trade_date"].iloc[0]
        window_start = grp["time_str"].min()
        ticker       = ticker_map.get(sec_id, str(sec_id))
        max_conc     = grp["bid_conc"].max()
        max_bid_size = grp["bid_size_level01"].max()
        duration_min = len(grp)

        remark = (
            f"{ticker}: bid level-1 concentration reached {max_conc:.1%} of total bid depth "
            f"(max L1 size {max_bid_size:,.0f} shares) over {duration_min} minutes from {window_start}. "
            f"Normal market has depth spread across levels; concentration >85% at L1 "
            f"suggests a single large order artificially supporting the bid."
        )
        severity = "HIGH" if max_conc > 0.95 or duration_min >= 8 else "MEDIUM"
        alerts.append({
            "sec_id":             sec_id,
            "trade_date":         trade_date,
            "time_window_start":  window_start,
            "anomaly_type":       "bid_level1_concentration_spike",
            "severity":           severity,
            "remarks":            remark,
        })
    return alerts


def detect_marking_the_close(trades: pd.DataFrame, ohlcv: pd.DataFrame, ticker_map: dict) -> list:
    """
    Detect concentrated selling/buying in the final 10 minutes of the session
    that accounts for an unusually large fraction of the day's total volume.
    """
    alerts = []
    filled = trades[trades["order_status"] == "FILLED"].copy()
    filled["trade_date"] = filled["timestamp"].dt.date.astype(str)
    filled["hour"]   = filled["timestamp"].dt.hour
    filled["minute"] = filled["timestamp"].dt.minute

    # Close window: 15:50 onwards
    in_close = (filled["hour"] == CLOSE_WINDOW_START_HOUR) & \
               (filled["minute"] >= CLOSE_WINDOW_START_MIN) | \
               (filled["hour"] >= 16)

    day_vol   = filled.groupby(["sec_id", "trade_date"])["quantity"].sum()
    close_vol = filled[in_close].groupby(["sec_id", "trade_date"])["quantity"].sum()
    ratio     = (close_vol / day_vol).dropna()

    for (sec_id, trade_date), pct in ratio.items():
        if pct < CLOSE_PCT_THRESH:
            continue
        ticker       = ticker_map.get(sec_id, str(sec_id))
        close_trades = filled[
            in_close &
            (filled["sec_id"] == sec_id) &
            (filled["trade_date"] == trade_date)
        ]
        window_start = close_trades["timestamp"].dt.strftime("%H:%M:%S").min()
        traders_inv  = close_trades["trader_id"].unique().tolist()
        total_qty    = int(close_trades["quantity"].sum())
        sides        = close_trades["side"].value_counts().to_dict()
        avg_price    = close_trades["price"].mean()

        # Compare to OHLCV close
        ohlcv_day = ohlcv[
            (ohlcv["sec_id"] == sec_id) &
            (ohlcv["trade_date"].dt.date.astype(str) == trade_date)
        ]
        close_px = ohlcv_day["close"].iloc[-1] if len(ohlcv_day) else None
        px_note  = f" OHLCV close ${close_px:.2f}." if close_px else ""

        remark = (
            f"{ticker}: {pct:.1%} of the day's filled volume ({total_qty:,} shares) "
            f"concentrated in the close window from {window_start}.{px_note} "
            f"Avg trade price ${avg_price:.2f}. "
            f"Traders: {', '.join(traders_inv)}. Side breakdown: {sides}. "
            f"Concentrated end-of-session activity may indicate marking the close."
        )
        severity = "HIGH" if pct > 0.90 else "MEDIUM"
        alerts.append({
            "sec_id":             sec_id,
            "trade_date":         trade_date,
            "time_window_start":  window_start,
            "anomaly_type":       "marking_the_close",
            "severity":           severity,
            "remarks":            remark,
        })
    return alerts


def detect_abnormal_volume_days(trades: pd.DataFrame, ohlcv: pd.DataFrame, ticker_map: dict) -> list:
    """
    Days where OHLCV volume is >3σ above the rolling 15-day baseline,
    corroborated by large individual trade activity.
    """
    alerts = []
    ohlcv_s = ohlcv.sort_values(["sec_id", "trade_date"])
    ohlcv_s["vol_15d_mean"] = ohlcv_s.groupby("sec_id")["volume"].transform(
        lambda x: x.shift(1).rolling(15, min_periods=5).mean()
    )
    ohlcv_s["vol_15d_std"] = ohlcv_s.groupby("sec_id")["volume"].transform(
        lambda x: x.shift(1).rolling(15, min_periods=5).std()
    )
    ohlcv_s["vol_z"] = (ohlcv_s["volume"] - ohlcv_s["vol_15d_mean"]) / \
                       ohlcv_s["vol_15d_std"].replace(0, np.nan)

    event_week = ohlcv_s[ohlcv_s["trade_date"] >= "2026-02-17"]
    high_vol   = event_week[event_week["vol_z"] > 3]

    filled = trades[trades["order_status"] == "FILLED"].copy()
    filled["trade_date"] = filled["timestamp"].dt.date.astype(str)

    for _, row in high_vol.iterrows():
        sec_id     = row["sec_id"]
        trade_date = row["trade_date"].strftime("%Y-%m-%d")
        vol_z      = row["vol_z"]
        ticker     = ticker_map.get(sec_id, str(sec_id))

        day_trades = filled[
            (filled["sec_id"] == sec_id) &
            (filled["trade_date"] == trade_date)
        ]
        if day_trades.empty:
            continue

        traders     = day_trades["trader_id"].unique().tolist()
        max_single  = day_trades["quantity"].max()
        total_qty   = day_trades["quantity"].sum()
        window_start = day_trades["timestamp"].dt.strftime("%H:%M:%S").min()

        remark = (
            f"{ticker}: daily volume {row['volume']:,} shares "
            f"({vol_z:.1f}σ above 15-day baseline of {row['vol_15d_mean']:,.0f}). "
            f"Trade-level: {len(day_trades)} fills, total {total_qty:,} shares, "
            f"largest single order {max_single:,}. "
            f"Active traders: {', '.join(traders)}. "
            f"Elevated volume may signal informed trading or coordinated activity."
        )
        severity = "HIGH" if vol_z > 6 else "MEDIUM"
        alerts.append({
            "sec_id":             sec_id,
            "trade_date":         trade_date,
            "time_window_start":  window_start,
            "anomaly_type":       "abnormal_daily_volume",
            "severity":           severity,
            "remarks":            remark,
        })
    return alerts


def detect_depth_ratio_anomalies(df: pd.DataFrame, ticker_map: dict) -> list:
    """
    Extreme depth ratio at top of book — bid L1 size >> ask L1 size or vice versa.
    Flags moments where a single side is massively over-represented at best price.
    """
    alerts = []
    extreme = df[
        (df["depth_ratio_z"].abs() > DEPTH_RATIO_Z) &
        df["depth_ratio"].notna() &
        (df["depth_ratio"] > 0)
    ].copy()

    if extreme.empty:
        return alerts

    extreme = extreme.sort_values(["sec_id", "timestamp"])
    extreme["run_id"] = (
        extreme.groupby("sec_id")["timestamp"]
        .transform(lambda x: (x.diff() > pd.Timedelta("3min")).cumsum())
    )

    for (sec_id, run_id), grp in extreme.groupby(["sec_id", "run_id"]):
        if len(grp) < 2:
            continue
        trade_date   = grp["trade_date"].iloc[0]
        window_start = grp["time_str"].min()
        ticker       = ticker_map.get(sec_id, str(sec_id))
        max_z        = grp["depth_ratio_z"].abs().max()
        max_ratio    = grp["depth_ratio"].max()
        duration_min = len(grp)

        direction = "bid far exceeds ask" if grp["depth_ratio_z"].mean() > 0 else "ask far exceeds bid"
        remark = (
            f"{ticker}: depth ratio at L1 reached {max_ratio:.1f}x ({direction}), "
            f"z-score {max_z:.1f}σ above 60-min baseline. "
            f"Persisted {duration_min} minutes from {window_start}. "
            f"Extreme top-of-book asymmetry consistent with spoofing or layering."
        )
        severity = "HIGH" if max_z > 7 or duration_min >= 8 else "MEDIUM"
        alerts.append({
            "sec_id":             sec_id,
            "trade_date":         trade_date,
            "time_window_start":  window_start,
            "anomaly_type":       "extreme_depth_ratio",
            "severity":           severity,
            "remarks":            remark,
        })
    return alerts


def cluster_and_filter(df: pd.DataFrame) -> pd.DataFrame:
    """
    Apply DBSCAN to OBI/spread/depth features. Return only anomalous windows
    (cluster -1) that pass a minimum threshold to reduce noise.
    """
    feats = ["obi_z", "spread_bps_z", "depth_ratio_z", "bid_conc_z"]
    valid = df[feats].dropna()
    if len(valid) < DBSCAN_MIN_SAMPLES * 2:
        return pd.DataFrame()

    scaler = StandardScaler()
    X = scaler.fit_transform(valid)
    labels = DBSCAN(eps=DBSCAN_EPS, min_samples=DBSCAN_MIN_SAMPLES).fit_predict(X)
    anomaly_idx = valid.index[labels == -1]
    anomalies = df.loc[anomaly_idx].copy()
    anomalies["dbscan_label"] = -1

    # Only keep windows with at least one extreme feature
    mask = (
        (anomalies["obi_z"].abs() > 3) |
        (anomalies["spread_bps_z"].abs() > 3) |
        (anomalies["depth_ratio_z"].abs() > 3) |
        (anomalies["bid_conc_z"].abs() > 3)
    )
    return anomalies[mask]


def generate_dbscan_alerts(df: pd.DataFrame, ticker_map: dict) -> list:
    """Run DBSCAN per ticker and generate alerts for anomalous windows."""
    alerts = []
    for sec_id, grp in df.groupby("sec_id"):
        ticker    = ticker_map.get(sec_id, str(sec_id))
        anomalies = cluster_and_filter(grp)
        if anomalies.empty:
            continue

        # Group anomalies into contiguous windows
        anomalies = anomalies.sort_values("timestamp")
        anomalies["run_id"] = (
            anomalies["timestamp"].diff() > pd.Timedelta("5min")
        ).cumsum()

        for run_id, run in anomalies.groupby("run_id"):
            if len(run) < 2:
                continue
            trade_date   = run["trade_date"].iloc[0]
            window_start = run["time_str"].min()
            max_obi_z    = run["obi_z"].abs().max()
            max_spr_z    = run["spread_bps_z"].abs().max()
            duration_min = len(run)

            remark = (
                f"{ticker}: DBSCAN identified {duration_min}-minute anomalous window "
                f"starting {window_start}. Max OBI z-score {max_obi_z:.1f}σ, "
                f"spread z-score {max_spr_z:.1f}σ. "
                f"Multi-feature deviation from baseline; consistent with structured manipulation."
            )
            alerts.append({
                "sec_id":             sec_id,
                "trade_date":         trade_date,
                "time_window_start":  window_start,
                "anomaly_type":       "multi_feature_order_book_anomaly",
                "severity":           "MEDIUM",
                "remarks":            remark,
            })
    return alerts


# ── Main ───────────────────────────────────────────────────────────────────────

def run(data_dir: str = DATA_DIR, output_file: str = OUTPUT_FILE):
    t0 = time.time()

    mkt, ohlcv, trades = load_data(data_dir)
    ticker_map = build_ticker_map(ohlcv)

    print("\nComputing order book features …")
    mkt_feat = compute_order_book_features(mkt)

    print("Running detection modules …")
    all_alerts = []
    all_alerts += detect_spoofing_from_trades(trades, ticker_map)
    all_alerts += detect_obi_concentration(mkt_feat, ticker_map)
    all_alerts += detect_spread_anomalies(mkt_feat, ticker_map)
    all_alerts += detect_bid_concentration_spike(mkt_feat, ticker_map)
    all_alerts += detect_marking_the_close(trades, ohlcv, ticker_map)
    all_alerts += detect_abnormal_volume_days(trades, ohlcv, ticker_map)
    all_alerts += detect_depth_ratio_anomalies(mkt_feat, ticker_map)
    all_alerts += generate_dbscan_alerts(mkt_feat, ticker_map)

    elapsed = time.time() - t0
    print(f"\nTotal raw alerts: {len(all_alerts)}")

    # ── Deduplicate ──
    # Keep the highest severity alert per (sec_id, trade_date, anomaly_type)
    df_alerts = pd.DataFrame(all_alerts)
    sev_order = {"HIGH": 0, "MEDIUM": 1, "LOW": 2}
    df_alerts["_sev_rank"] = df_alerts["severity"].map(sev_order)
    df_alerts = (
        df_alerts
        .sort_values(["sec_id", "trade_date", "anomaly_type", "_sev_rank"])
        .drop_duplicates(subset=["sec_id", "trade_date", "anomaly_type"])
        .drop(columns=["_sev_rank"])
        .reset_index(drop=True)
    )

    df_alerts.insert(0, "alert_id", range(1, len(df_alerts) + 1))
    df_alerts["time_to_run"] = round(elapsed, 2)

    # Re-order columns as per spec
    cols = ["alert_id", "sec_id", "trade_date", "time_window_start",
            "anomaly_type", "severity", "remarks", "time_to_run"]
    df_alerts = df_alerts[cols]

    df_alerts.to_csv(output_file, index=False)
    print(f"\nAlerts written to {output_file}  ({len(df_alerts)} rows)")
    print(f"Time to run: {elapsed:.2f}s")
    print("\nAlert summary:")
    print(df_alerts.groupby(["anomaly_type", "severity"]).size().to_string())
    return df_alerts


# ── Entry point ────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    alerts_df = run()
    print("\nSample alerts:")
    print(alerts_df[["alert_id", "sec_id", "trade_date",
                      "anomaly_type", "severity"]].to_string())
```
