```python
"""
Problem 2 — Insider Trading Signal (Equity, Bonus)
====================================================
Scrapes SEC EDGAR for 8-K filings across the ticker universe,
builds a news event timeline, and cross-references it against trade data
to surface pre-announcement insider trading signals.

Outputs: p2_signals.csv

Usage (JupyterLab):
    Run all cells top-to-bottom, or:
    %run problem2_insider_trading.py
    from problem2_insider_trading import run; run()
"""

import time
import warnings
import numpy as np
import pandas as pd
import requests
from time import sleep

warnings.filterwarnings("ignore")

# ── Configuration ──────────────────────────────────────────────────────────────
DATA_DIR    = 'data_dir/equity'
OUTPUT_FILE = "p2_signals.csv"

EDGAR_URL   = "https://efts.sec.gov/LATEST/search-index"

# Date range to scrape (covers 15-day baseline + event week)
EDGAR_START = "2026-01-01"
EDGAR_END   = "2026-02-20"

# Signal thresholds
VOLUME_Z_THRESHOLD   = 2.0   # flag if volume z-score on T-1 or T-2 > this
DRIFT_SIGMA_THRESHOLD = 2.0  # flag if cumulative T-5:T-1 drift > 2σ above baseline
TRADER_ZSCORE_THRESH  = 2.5  # flag individual trader quantity z-score
BASELINE_DAYS         = 15   # rolling window for baseline stats

# Event type keyword map
EVENT_KEYWORDS = {
    "merger":      ["merger", "acquisition", "acquired", "takeover",
                    "combine", "definitive agreement", "strategic transaction"],
    "earnings":    ["earnings", "revenue", "quarterly results", "guidance",
                    "eps", "net income", "financial results"],
    "leadership":  ["ceo", "chief executive", "resign", "appoint",
                    "board", "director", "officer", "president"],
    "restatement": ["restate", "restatement", "correction",
                    "material weakness", "error"],
    "bankruptcy":  ["bankruptcy", "chapter 11", "restructuring",
                    "insolvency", "liquidation"],
    "contract":    ["contract", "agreement", "partnership", "collaboration",
                    "awarded", "license"],
}


# ── Data Loading ───────────────────────────────────────────────────────────────

def load_data(data_dir: str):
    """Load OHLCV and trade data CSVs."""
    print("Loading equity data …")
    ohlcv  = pd.read_csv(f"{data_dir}/ohlcv.csv",      parse_dates=["trade_date"])
    trades = pd.read_csv(f"{data_dir}/trade_data.csv",  parse_dates=["timestamp"])
    print(f"  ohlcv       : {len(ohlcv):,} rows | "
          f"{ohlcv['trade_date'].min().date()} – {ohlcv['trade_date'].max().date()}")
    print(f"  trade_data  : {len(trades):,} rows")
    return ohlcv, trades


def build_ticker_map(ohlcv: pd.DataFrame) -> pd.DataFrame:
    """Return unique (sec_id, ticker, name) rows."""
    return ohlcv[["sec_id", "ticker", "name"]].drop_duplicates().reset_index(drop=True)


# ── EDGAR Scraping ─────────────────────────────────────────────────────────────

def classify_event(text: str) -> str:
    """Keyword-based event type classifier."""
    t = text.lower()
    for event_type, keywords in EVENT_KEYWORDS.items():
        if any(kw in t for kw in keywords):
            return event_type
    return "other"


def fetch_8k_for_ticker(ticker: str, start_date: str, end_date: str,
                        session: requests.Session) -> list:
    """Fetch 8-K filings from EDGAR for a single ticker."""
    params = {
        "q":         f'"{ticker}"',
        "forms":     "8-K",
        "dateRange": "custom",
        "startdt":   start_date,
        "enddt":     end_date,
    }
    try:
        resp = session.get(EDGAR_URL, params=params, timeout=15)
        resp.raise_for_status()
        data = resp.json()
    except Exception as e:
        print(f"    EDGAR request failed for {ticker}: {e}")
        return []

    results = []
    hits = data.get("hits", {}).get("hits", [])
    for hit in hits:
        src = hit.get("_source", {})
        entity = src.get("entity_name", "")
        file_date = src.get("file_date", "")
        file_path = src.get("file_path", "")
        form_type = src.get("form_type", "")
        # Grab display_names / description for headline
        # display_names can be a list of dicts OR a list of strings depending on EDGAR version
        display_names = src.get("display_names", [])
        if display_names:
            first = display_names[0]
            if isinstance(first, dict):
                headline_raw = first.get("name", entity)
            else:
                headline_raw = str(first)
        else:
            headline_raw = entity
        # Try to get a richer description
        description = src.get("period_of_report", "")
        results.append({
            "ticker":    ticker,
            "entity":    entity,
            "file_date": file_date,
            "form_type": form_type,
            "headline":  headline_raw,
            "filing_url": f"https://www.sec.gov{file_path}" if file_path else "",
        })
    return results


def scrape_edgar(tickers: list, start_date: str, end_date: str) -> pd.DataFrame:
    """
    Scrape SEC EDGAR for 8-K filings for all tickers.
    Falls back gracefully when the API is unavailable.
    """
    print(f"\nScraping EDGAR for {len(tickers)} tickers ({start_date} – {end_date}) …")
    session = requests.Session()
    session.headers.update({"User-Agent": "surveillance-hackathon contact@example.com"})

    all_results = []
    for i, ticker in enumerate(tickers):
        results = fetch_8k_for_ticker(ticker, start_date, end_date, session)
        if results:
            print(f"  [{i+1:02d}/{len(tickers)}] {ticker}: {len(results)} filings")
        all_results.extend(results)
        sleep(0.35)   # polite delay

    if not all_results:
        print("  No EDGAR results retrieved (network unavailable).")
        return pd.DataFrame(columns=["ticker", "entity", "file_date", "form_type",
                                     "headline", "filing_url", "event_type"])

    df = pd.DataFrame(all_results)
    df["file_date"]  = pd.to_datetime(df["file_date"], errors="coerce")
    df               = df.dropna(subset=["file_date"])
    df["event_type"] = df["headline"].apply(classify_event)
    df               = df.sort_values("file_date").reset_index(drop=True)
    print(f"  Total filings: {len(df)}")
    print(f"  Event types: {df['event_type'].value_counts().to_dict()}")
    return df


# ── Signal Detection ───────────────────────────────────────────────────────────

def build_ohlcv_baseline(ohlcv: pd.DataFrame) -> pd.DataFrame:
    """
    Compute rolling 15-day volume and return baselines per sec_id.
    Shift by 1 so the event-day baseline is computed from pre-event data only.
    """
    df = ohlcv.sort_values(["sec_id", "trade_date"]).copy()
    df["close_ret"] = df.groupby("sec_id")["close"].pct_change()

    grp = df.groupby("sec_id")
    df["vol_15d_mean"] = grp["volume"].transform(
        lambda x: x.shift(1).rolling(BASELINE_DAYS, min_periods=5).mean()
    )
    df["vol_15d_std"] = grp["volume"].transform(
        lambda x: x.shift(1).rolling(BASELINE_DAYS, min_periods=5).std()
    )
    df["ret_15d_mean"] = grp["close_ret"].transform(
        lambda x: x.shift(1).rolling(BASELINE_DAYS, min_periods=5).mean()
    )
    df["ret_15d_std"] = grp["close_ret"].transform(
        lambda x: x.shift(1).rolling(BASELINE_DAYS, min_periods=5).std()
    )
    df["vol_z"] = (df["volume"] - df["vol_15d_mean"]) / df["vol_15d_std"].replace(0, np.nan)
    return df


def compute_pre_drift(ohlcv_base: pd.DataFrame, sec_id: int,
                      event_date: pd.Timestamp) -> dict:
    """
    Compute cumulative abnormal return from T-5 to T-1 relative to the
    rolling 15-day baseline.
    """
    sec_data = ohlcv_base[ohlcv_base["sec_id"] == sec_id].copy()
    pre_window = sec_data[
        (sec_data["trade_date"] >= event_date - pd.Timedelta(days=10)) &
        (sec_data["trade_date"] < event_date)
    ].tail(5)  # T-5 to T-1

    if len(pre_window) < 2:
        return {"pre_drift": np.nan, "vol_z_max": np.nan, "drift_flag": 0,
                "vol_flag": 0, "window_start": None}

    baseline_ret  = pre_window["ret_15d_mean"].mean()
    baseline_std  = pre_window["ret_15d_std"].mean()

    # Cumulative abnormal return
    car = (pre_window["close_ret"] - baseline_ret).sum()
    car_z = car / (baseline_std * np.sqrt(len(pre_window))) if baseline_std and baseline_std > 0 else 0

    vol_z_max = pre_window["vol_z"].max()

    drift_flag = 1 if abs(car_z) > DRIFT_SIGMA_THRESHOLD else 0
    vol_flag   = 1 if vol_z_max > VOLUME_Z_THRESHOLD else 0

    return {
        "pre_drift":      round(car, 6),
        "car_z":          round(car_z, 3),
        "vol_z_max":      round(vol_z_max, 3) if not np.isnan(vol_z_max) else np.nan,
        "drift_flag":     drift_flag,
        "vol_flag":       vol_flag,
        "window_start":   pre_window["trade_date"].min(),
    }


def find_suspicious_traders(trades: pd.DataFrame, sec_id: int,
                             event_date: pd.Timestamp) -> dict:
    """
    Identify individual traders with abnormally large positions in T-5 to T-1.
    Compares each trader's activity in the window against their overall baseline.
    """
    filled = trades[
        (trades["sec_id"] == sec_id) &
        (trades["order_status"] == "FILLED")
    ].copy()

    if filled.empty:
        return {}

    filled["trade_date"] = filled["timestamp"].dt.date

    # Pre-window: T-5 to T-1
    t_minus_5 = (event_date - pd.Timedelta(days=7)).date()
    t_minus_1 = (event_date - pd.Timedelta(days=1)).date()

    pre_filled = filled[
        (filled["trade_date"] >= t_minus_5) &
        (filled["trade_date"] <= t_minus_1)
    ]

    if pre_filled.empty:
        return {}

    # Baseline: overall daily quantity per trader
    trader_baseline = (
        filled.groupby(["trader_id", "trade_date"])["quantity"]
        .sum().reset_index()
        .groupby("trader_id")["quantity"]
        .agg(["mean", "std"]).rename(columns={"mean": "baseline_mean", "std": "baseline_std"})
    )

    # Pre-window daily quantity per trader
    pre_daily = pre_filled.groupby(["trader_id", "trade_date"])["quantity"].sum().reset_index()
    pre_daily = pre_daily.merge(trader_baseline, on="trader_id", how="left")
    pre_daily["qty_z"] = (
        (pre_daily["quantity"] - pre_daily["baseline_mean"]) /
        pre_daily["baseline_std"].replace(0, np.nan)
    )

    suspicious = pre_daily[pre_daily["qty_z"] > TRADER_ZSCORE_THRESH]
    if suspicious.empty:
        return {}

    result = {}
    for trader_id, grp in suspicious.groupby("trader_id"):
        result[trader_id] = {
            "max_qty":      int(grp["quantity"].max()),
            "max_qty_z":    round(grp["qty_z"].max(), 2),
            "dates":        grp["trade_date"].tolist(),
            "side":         pre_filled[pre_filled["trader_id"] == trader_id]["side"].mode().iloc[0]
                            if len(pre_filled[pre_filled["trader_id"] == trader_id]) else "?",
        }
    return result


def build_signals(filings: pd.DataFrame, ohlcv: pd.DataFrame,
                  trades: pd.DataFrame, ticker_map: pd.DataFrame) -> pd.DataFrame:
    """
    For each (ticker, filing_date) pair, compute:
      - Pre-announcement drift (cumulative abnormal return T-5 to T-1)
      - Abnormal volume z-scores
      - Trade-level suspicious trader evidence
    """
    if filings.empty:
        print("No EDGAR filings to process. Building signals from volume anomalies only.")
        return _build_volume_only_signals(ohlcv, trades, ticker_map)

    ohlcv_base = build_ohlcv_baseline(ohlcv)
    tm_dict = dict(zip(ticker_map["ticker"], ticker_map["sec_id"]))

    rows = []
    seen = set()

    for _, filing in filings.iterrows():
        ticker     = filing["ticker"]
        event_date = filing["file_date"]

        if not isinstance(event_date, pd.Timestamp):
            continue

        sec_id = tm_dict.get(ticker)
        if sec_id is None:
            continue

        key = (sec_id, event_date.date())
        if key in seen:
            continue
        seen.add(key)

        drift_stats       = compute_pre_drift(ohlcv_base, sec_id, event_date)
        suspicious_traders = find_suspicious_traders(trades, sec_id, event_date)

        pre_drift_flag = max(drift_stats["drift_flag"], drift_stats["vol_flag"])

        # Build remarks
        remarks_parts = []
        if drift_stats["vol_flag"]:
            remarks_parts.append(
                f"Volume z-score {drift_stats['vol_z_max']:.1f}σ in T-5:T-1 window "
                f"(threshold {VOLUME_Z_THRESHOLD}σ)."
            )
        if drift_stats["drift_flag"]:
            remarks_parts.append(
                f"Cumulative abnormal return {drift_stats['pre_drift']:.4f} "
                f"({drift_stats['car_z']:.1f}σ above 15-day baseline)."
            )
        if suspicious_traders:
            for tid, info in suspicious_traders.items():
                remarks_parts.append(
                    f"Trader {tid} qty z-score {info['max_qty_z']:.1f}σ "
                    f"({info['max_qty']:,} shares {info['side']}) in pre-window."
                )
        if not remarks_parts:
            remarks_parts.append("No significant pre-announcement signal detected.")

        rows.append({
            "sec_id":                  sec_id,
            "ticker":                  ticker,
            "event_date":              event_date.strftime("%Y-%m-%d"),
            "event_type":              filing["event_type"],
            "headline":                str(filing["headline"])[:200],
            "source_url":              filing["filing_url"],
            "pre_drift_flag":          pre_drift_flag,
            "suspicious_window_start": (
                drift_stats["window_start"].strftime("%Y-%m-%d")
                if drift_stats["window_start"] is not None else ""
            ),
            "remarks":                 " ".join(remarks_parts),
        })

    df_sig = pd.DataFrame(rows)
    if df_sig.empty:
        return pd.DataFrame()

    return df_sig.sort_values("pre_drift_flag", ascending=False).reset_index(drop=True)


def _build_volume_only_signals(ohlcv: pd.DataFrame, trades: pd.DataFrame,
                                ticker_map: pd.DataFrame) -> pd.DataFrame:
    """
    Fallback: when EDGAR is unavailable, flag days with high volume z-scores
    during the event week as candidates for pre-announcement activity.
    """
    print("\nFallback: using volume-only signals …")
    ohlcv_base = build_ohlcv_baseline(ohlcv)
    event_week = ohlcv_base[ohlcv_base["trade_date"] >= "2026-02-17"].copy()
    high_vol   = event_week[event_week["vol_z"] > VOLUME_Z_THRESHOLD].copy()

    tm_dict = dict(zip(ticker_map["sec_id"], ticker_map["ticker"]))
    rows = []
    for _, row in high_vol.iterrows():
        sec_id     = row["sec_id"]
        trade_date = row["trade_date"]
        vol_z      = row["vol_z"]
        ticker     = tm_dict.get(sec_id, str(sec_id))

        # Check for suspicious traders
        suspicious_traders = find_suspicious_traders(
            trades, sec_id, trade_date + pd.Timedelta(days=1)
        )
        trader_note = ""
        if suspicious_traders:
            for tid, info in suspicious_traders.items():
                trader_note += (
                    f" Trader {tid}: {info['max_qty']:,} shares {info['side']}, "
                    f"z-score {info['max_qty_z']:.1f}σ."
                )

        remarks = (
            f"Elevated volume on {trade_date.date()}: z-score {vol_z:.1f}σ above "
            f"15-day baseline of {row['vol_15d_mean']:,.0f} shares "
            f"(actual {row['volume']:,})." + trader_note +
            " Flagged for potential pre-announcement informed trading."
        )

        rows.append({
            "sec_id":                  sec_id,
            "ticker":                  ticker,
            "event_date":              trade_date.strftime("%Y-%m-%d"),
            "event_type":              "abnormal_volume",
            "headline":                f"{ticker} abnormal volume {vol_z:.1f}σ above baseline",
            "source_url":              "",
            "pre_drift_flag":          1 if vol_z > 3 else 0,
            "suspicious_window_start": trade_date.strftime("%Y-%m-%d"),
            "remarks":                 remarks,
        })

    return pd.DataFrame(rows).sort_values("pre_drift_flag", ascending=False).reset_index(drop=True)


# ── Main ───────────────────────────────────────────────────────────────────────

def run(data_dir: str = DATA_DIR, output_file: str = OUTPUT_FILE):
    t0 = time.time()

    ohlcv, trades = load_data(data_dir)
    ticker_map    = build_ticker_map(ohlcv)
    tickers       = ticker_map["ticker"].tolist()

    # ── Scrape EDGAR ──
    filings = scrape_edgar(tickers, EDGAR_START, EDGAR_END)

    # ── Build signals ──
    print("\nBuilding pre-announcement signals …")
    signals = build_signals(filings, ohlcv, trades, ticker_map)

    elapsed = time.time() - t0

    if signals.empty:
        print("No signals generated.")
        return pd.DataFrame()

    signals["time_to_run"] = round(elapsed, 2)

    # Re-order columns per spec
    cols_order = [
        "sec_id", "event_date", "event_type", "headline", "source_url",
        "pre_drift_flag", "suspicious_window_start", "remarks", "time_to_run",
    ]
    signals = signals[[c for c in cols_order if c in signals.columns]]
    signals.to_csv(output_file, index=False)

    flagged = signals[signals["pre_drift_flag"] == 1]
    print(f"\nSignals written to {output_file}  ({len(signals)} rows total)")
    print(f"Flagged pre-drift signals: {len(flagged)}")
    print(f"Time to run: {elapsed:.2f}s")

    if not flagged.empty:
        print("\nTop flagged signals:")
        pd.set_option("display.max_colwidth", 80)
        print(
            flagged[["sec_id", "event_date", "event_type", "remarks"]]
            .head(10)
            .to_string(index=False)
        )

    return signals


# ── Entry point ────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    signals_df = run()
```
